# **FUNDAMENTALS OF NUMPY, PANDAS AND MATPLOTLIB**

## **NUMPY**

```python
import numpy as np 
```

```python
#Create a numpy array
a=np.array([1,2,3,4,5])
print(a)
```

```python
#Create array with zeroes
print(np.zeros((3,2)))
```

```python
#Create array with constant
print(np.full((2,3),7))
```

```python
#Create array with random value
print(np.random.random((2,2)))
```

```python
#Create array with sequence of integers
x=np.arange(4, 28, 4)
print(x)
```

```python
#Reshape an array
y=x.reshape(2,3)
print(y)
```

```python
#Sort an array
print(np.sort(y))
```

```python
#Size and shape of array
print(y.ndim)
print(y.shape)
print(y.size)
print(y.dtype)
```

```python
#Flatten an array
y=y.flatten()
print(y)
```

```python
#Convert array datatype
y2=y.astype("f")
print(y2,y2.dtype)
```

```python
#Scalar operation
x=x*2
print(x)
y=y+1
print(y)
```

```python
#Vector operation
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x.dot(y))
```

```python
#Joining two arrays
print(np.hstack((x,y)))
print(np.vstack((x,y)))
print(np.dstack((x,y)))
```

## **PANDAS**

```python
import pandas as pd
```

```python
#Create series with array
a=np.array([1,2,3,4,5])
b=pd.Series(a)
print(b)
```

```python
#Create series with dictionary
a={1:'a',2:'b',3:'c'}
b=pd.Series(a)
print(b)
```

```python
#Series using scalar
x = pd.Series(4, index=[0, 1, 2, 3])  
print("\nSeries using scalar\n")
print (x) 
```

```python
#Create series with index
print(pd.Series(7, index=[1,2,3,4]))
```

```python
#Size and shape
print(b.ndim)
print(b.size)
print(b.index)
print(b.values)
```

```python
emp = pd.Series(['Parker', 'John', 'Smith', 'William'])   
id =pd.Series([102, 107, 109, 114])     
frame = { 'Emp': emp, 'ID': id }   
result = pd.DataFrame(frame) 
print("\nSeries to Data frame\n")
print(result)
```

```python
left = pd.DataFrame({  
   'id':[1,2,3,4,5],  
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],  
   'sub':['sub1','sub2','sub4','sub6','sub5']})  
right = pd.DataFrame({  
    'id':[1,2,3,4,5],  
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],  
   'sub':['sub2','sub4','sub3','sub6','sub5']})  
print(left,'\n')
print(right,'\n')
print(pd.merge(left,right,on='id'))
```

## **MATPLOTLIB**

```python
import matplotlib.pyplot as plt
```

```python
#Plot using Matplotlib
x = np.array([0,1,2,3,4])
y = np.array([0,5,2,8,10])
plt.plot(x,y)
plt.show()
```

```python
#Plot with default x values
y = np.array([5,9,8,11])
plt.plot(y)
plt.show()
```

```python
#Grid plot
y = np.array([5,9,8,11])
plt.plot(y)
plt.grid()
plt.show()
```

```python
#Marker in plot
x = np.array([0,1,2,3,4])
y = np.array([0,5,2,8,10])
plt.plot(x,y,marker='o')
plt.show()
```

```python
#Plot without lines
plt.plot(x,y,'s')
plt.show()
```

```python
#Plot using linestyle and Subplot
plt.subplot(2,1,1)
plt.plot(x,y,linestyle='dotted')
plt.show()
plt.subplot(2,1,2)
plt.plot(x,y,linestyle='dashed')

plt.show()
```

```python
#colour of line
plt.plot(x,y,color='r')
plt.show()
```

```python
#Colour and linestyle
plt.plot(x,y,'s:k')
plt.show()
```

```python
#Scatter plot with colour and size
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
colors = np.random.randint(15, size=(15))
sizes = 10 * np.random.randint(15, size=(15))
plt.scatter(x,y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

```python
#Pie chart
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```
