# **Create a linear regression model using keras API**

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense

# Generate a synthetic dataset
np.random.seed(0)
X = np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Create the linear regression model using Keras
model = tf.keras.Sequential([tf.keras.layers.Dense(1, input_shape=(1,))])

# Compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')

# Train the model
history = model.fit(X,y, epochs=100, batch_size=1, verbose=0)

# Make predictions
pred = model.predict(X)

# Plot the results
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, pred, color='red', label='Predicted')
plt.title('Linear Regression with Keras')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

print("WEIGHTS:", model.layers[0].get_weights()[0])
print("BIAS:", model.layers[0].get_weights()[1])
```
